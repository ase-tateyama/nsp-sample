Matplotlibとは、グラフを可視化するためのライブラリ<br>

### ■Matoplotlibの使用
Matoplotlibを利用するためにインポートする。<br>

```python
# numpy モジュールも併せてインポート
import numpy as np
import matplotlib.pyplot as plt
# 「%matplotlib inline」は、notebook内でグラフを表示するために必要
%matplotlib inline
```

### 線グラフ
pyplot モジュールの plot() 関数を用いて、リストの要素の数値をy軸の値としてグラフを描画する。<br>
y軸の値に対応するx軸の値は、リストの各要素のインデックスとなっている。

```python
# データ
data =[0, 1, 4, 9, 16]

# plot
plt.plot(data)
```

plot() 関数では、x, y の両方の軸の値を引数に渡すこともできる。<br>
リストxとリストyを引数とすると、各リストの値の位置に点を打ち、各点が線でつながる。<br>
※リストの代わりにNumPyライブラリの配列でも同じ結果が得られる。

```python
# データ
x =[0, 1, 2, 3, 4]
y =[16, 15, 13, 9, 1]

# plot
plt.plot(x,y);
```

グラフを複数まとめて表示することも可能。<br>
複数のグラフを表示すると、線ごとに異なる色が自動で割り当てられる。<br>
また、引数で線の色等を指定可能。<br>
・linestyle：グラフの線の色<br>
・color：形状<br>
・marker：データポイントのマーカの種類

```python
# データ
data1 =[0, 1, 4, 9, 16]
x1 =[0, 1, 2, 3, 4]
y1 =[0, 4, 8, 12, 16]

data2 =[0, 1, 2, 3, 4]
x2 =[0, 1, 2, 3, 4]
y2 =[0, 7, 12, 15, 16]

# plot
plt.plot(data1);
plt.plot(x1, y1)
plt.plot(data2, linestyle='--', color='red', marker='o')
plt.plot(x2, y2, linestyle=':', color='green', marker='*')
```

plot() 関数のlabel引数にグラフの各線の凡例を文字列として渡し、legend() 関数を呼ぶことで、グラフ内に凡例を表示できる。<br>
（legend() 関数の loc 引数で凡例を表示する位置を指定することが可能）<br>
<br>
pyplotモジュールでは、グラフのタイトルと各軸のラベルを指定して表示することができる。<br>
タイトル：title() 関数<br>
x軸のラベル：xlabel() 関数<br>
y軸のラベル：ylabel() 関数<br>
※文字列を渡して指定する。<br>
<br>
grid() 関数を用いるとグリッドを併せて表示することができる。

```python
# データ
data =[0, 1, 2, 3, 4]
x =[0, 1, 2, 3, 4]
y =[0, 1, 4, 9, 16]

# plot
plt.plot(data, linestyle='--', color='red', marker='o', label='label1')
plt.plot(x, y, linestyle=':', color='green', marker='*', label='label2')
plt.legend()

# グラフのタイトル
plt.title('title') 
#x軸のラベル
plt.xlabel('x')
#y軸のラベル
plt.ylabel('y') 
#グリッドの表示
plt.grid(True);
```

### 散布図
散布図は、pyplotモジュールのscatter()関数を用いて描画できる。<br>
リストxとリストyを引数として与えると、各リストの値の位置に点が打たれる。<br>
※配列でも可能<br>
<br>
プロットする点のマーカの色や形状は、線グラフの時と同様に、color, marker 引数で指定して変更することができる。<br>
また、s, alpha引数で、それぞれマーカの大きさと透明度を指定することができる。

```python
#　x軸の値となる配列
x = np.random.rand(30)
#　y軸の値となる配列
y = np.random.rand(30)

print(x, y)

# scatter
plt.scatter(x, y, s=50, alpha=0.2)
```

### 棒グラフ
棒グラフは、pyplot モジュールの bar() 関数を用いて描画できる。<br>

```python
# x軸上で棒の並ぶ位置となる配列（都市名）
x = np.array(['Sapporo', 'Sendai', 'Tokyo', 'Nagoya', 'Osaka', 'Fukuoka', 'Naha'])
#　グラフのy軸の値となる配列（各都市の降水確率）
y = np.array([10, 50, 20, 30, 20, 0, 80])

# bar
print(x, y)
plt.bar(x,y)
plt.ylim(0, 100)
```

barh() 関数を用いると横向きの棒グラフが描画できる。

```python
plt.barh(x,y)
plt.xlim(0, 100)
```

### ヒストグラム
ヒストグラムは、pyplotモジュールのhist()関数を用いて描画できる。<br>
hist()関数のbins引数でヒストグラムの箱（ビン）の数を指定する。

```python
# 正規分布に基づく1000個の数値の要素からなる配列
d = np.random.randn(1000)

# hist関数でヒストグラムを描画
plt.hist(d, bins=50);
```

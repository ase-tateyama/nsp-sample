## データ分割

### 元データを分析用データと予測対象データに分割する

※本来、予測対象データは分析用データとは別のデータとなるが、今回は元データの一部を予測対象データとします。

#### 予測対象データとして元データからランダムに30件を抽出

```python
df_pred = df.sample(n=30, random_state=0)
```

```python
df_pred.info()
```

#### 学習用データとして元データから予測対象データを除去

```python
df_train = df[~df['TEAM'].isin(df_pred['TEAM'])]
```

```python
df_train.info()
```

#### データを目的変数と説明変数に分離する

```python
# df_xxx_Xは目的変数を除去した説明変数のみのデータ
df_train_X = df_train.drop(['W'], axis=1)
df_pred_X = df_pred.drop(['W'], axis=1)
# df_xxx_yは目的変数のみのデータ
df_train_y = df_train['W']
df_pred_y = df_pred['W']
```

#### 説明変数のデータからカテゴリ変数を除去する

```python
df_train_X = df_train_X.drop(['TEAM'], axis=1)
df_pred_X = df_pred_X.drop(['TEAM'], axis=1)
```

```python
df_train_X.head()
```

```python
df_pred_X.head()
```

#### データの標準化

```python
from sklearn.preprocessing import StandardScaler

# 分析用データを標準化する
X_sc_df = StandardScaler()

# 分析用データからパラメータ(平均、分散)を算出する
X_sc_df.fit(df_train_X)
# 分析用データを変換
df_train_X_std = X_sc_df.transform(df_train_X)

# 予測対象データを標準化する
X_sc_df = StandardScaler()

# 予測対象データからパラメータ(平均、分散)を算出する
X_sc_df.fit(df_pred_X)
# 予測対象データを変換
df_pred_X_std = X_sc_df.transform(df_pred_X)

# 目的変数は一列しかないので標準化の式に当てはめて変換
df_train_y_std = (df_train_y - df_train_y.mean())/df_train_y.std()
df_pred_y_std = (df_pred_y - df_pred_y.mean())/df_pred_y.std()
```

## 重回帰分析のためのモデル作成、実行

### 必要な外部ライブラリのimport

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

### データを学習用データと検証用データに分割

```python
linreg = LinearRegression()
```

```python
#学習データと予測データに分割(学習データ:予測データ=7:3)
X_train_data,X_test_data,y_train_data,y_test_data = train_test_split(df_train_X_std,df_train_y_std,test_size=0.3)
```

### モデルの作成と実行

```python
# 線形回帰モデルへ当てはめる
linreg.fit(X_train_data,y_train_data)
```

```python
# 予測実行
test_pred_result = linreg.predict(X_test_data)
```

### 予測結果の逆標準化

```python
# 予測値
y_test_data.predict = test_pred_result * df_train_y.std() + df_train_y.mean()
# 正解
y_test_val = y_test_data * df_train_y.std() + df_train_y.mean()
```

### 予実結果の可視化

```python
import matplotlib.pyplot as plt
%matplotlib inline
plt.figure(figsize=(10,10))
plt.scatter(y_test_val, y_test_data.predict)
plt.xlim(0, 40)
plt.ylim(0, 40)
plt.xlabel("Actual Wins")
plt.ylabel("Predicted Wins")
plt.gca().set_aspect("equal", adjustable="box")

x = np.arange(0, 35)
y=x
plt.plot(x,y,color="red")
```

### 誤差の計測

```python
rmse = np.sqrt(sum((y_test_val - y_test_data.predict) ** 2) / len(y_test_val))
print("RMSE = ", rmse)
```

### 説明変数の影響度を確認

```python
coef_df = pd.DataFrame({"Coef":linreg.coef_, "Coef_abs":np.abs(linreg.coef_)})
coef_df.index = df_train_X.columns
coef_df = coef_df.sort_values(by="Coef_abs")
coef_df["Coef"].tail(20).plot(kind="barh", title="Coefficients", color="red")
coef_df["Coef"].tail(20)
```

### 予測対象データに対して重回帰分析を実施する

```python
y_pred_result = linreg.predict(df_pred_X_std)
```

### 説明変数の影響度を確認

```python
coef_all_df = pd.DataFrame({"Coef":linreg.coef_, "Coef_abs":np.abs(linreg.coef_)})
coef_all_df.index = df_train_X.columns
coef_all_df = coef_all_df.sort_values(by="Coef_abs")
coef_all_df["Coef"].tail(20).plot(kind="barh", title="Coefficients", color="red")
coef_all_df["Coef"].tail(20)
```
